PyCitySchools Analysis:
- Charter schools significantly outperformed District Schools in their Overall Passing Rates.
- Small and medium sized schools, regardless of their per student budgets significantly outperform large sized schools
- The top performing schools tend to be the ones that do not spend as much money on their students (budget/student size ratio).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete_20181003.csv"
student_data_to_load = "Resources/students_complete_20181003.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
grouped_school_spending = school_data_complete
grouped_school_spending.head()
```

```python
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
grouped_school_spending["Spending Ranges Per Student"] = (grouped_school_spending['budget'] /\
    grouped_school_spending['size'])
grouped_school_spending.head()
```

```python
grouped_school_spending["Spending Ranges Per Student"] = pd.cut(grouped_school_spending\
    ["Spending Ranges Per Student"],bins,labels=group_names)
grouped_school_spending
```

```python
grouped_school_data_spending = grouped_school_spending.loc\
    [:,["is_math_passed","is_reading_passed","Average Math Score","Average Reading Score",\
    "Spending Ranges Per Student"]].groupby("Spending Ranges Per Student")
grouped_mean = grouped_school_data_spending.mean()
grouped_mean["% Overall Passing Rate"] =(grouped_mean.loc[:,"is_math_passed"]+grouped_mean.loc\
    [:,"is_reading_passed"])/2
grouped_mean
```

Scores by School Spending:

```python
grouped_mean = grouped_mean.rename(columns={'is_math_passed':'% Passing Math',\
    'is_reading_passed':'% Passing Reading','size':'School Size'})
(grouped_mean * 100).apply(round)
```
